**Specifying which sheet should be read**

To specify which sheet should be read, pass the value to the sheetname option. We will simply pass 0, which is the index value of the first sheet in the Excel sheet

```python
df = pd.read_excel('IMDB.xlsx',sheet_name=0)
df.head()
```

**Read data from multiple sheets**

Excel dataset files come with data and multiple sheets. In fact this is one of the main reasons a lot of users prefer Excel over csv file. Luckily pandas supports the reading of data from multiple sheets.

**Finding out sheet names**

To find out the name of a sheet, pass the Excel file to the ExcelFile class and call the sheet_names properly on the resulting object. The class prints the sheet names from the Excel file as a list. If you want to read data from a sheet named data-movies, we will write codes about that.

```python
xls_file = pd.ExcelFile('IMDB.xlsx')
xls_file.sheet_names
```

Next we will call the parse method on the Excel file object we created earlier and pass in the sheet names we want to read. We then assign the result to 2 seperate DataFrame objects.

```python
df1 = xls_file.parse('movies')
df2 = xls_file.parse('by genre')
df1.head()
```

We now have our data from 2 sheets in 2 seperate Dataframes as shown in the above tabke

**Choose header or column labels**

Pandas will, by default set the column names or header to the values from the first non-blank row in teh Excel file. However, we can change this behavior.

```python
df = pd.read_excel('IMDB.xlsx',sheet_name=1, header=3)
df.head()
```

In the above table we are passing the value 3 to the header option, which tells the read_excel method to set the header names from index row 3.

**No header**

We can also tell read_excel to ignore the header and treat all rows as records. This is handy whenever Excel has no header row. To achieve this we set header to None.

```python
df = pd.read_excel('IMDB.xlsx',sheet_name=1, header=None)
df.head()
```

**Skipping rows at the begining**

to skip rows at the begining of a file, we simply set skiprows to the number of rows we want to skip

```python
df = pd.read_excel('IMDB.xlsx',sheet_name=1, skiprows=7)
df.head()
```

**Skipping rows at the end**

We will use the skip_footer.

```python
df = pd.read_excel('IMDB.xlsx',sheet_name=1, skipfooter=10)
df.head()
```

**Choosing colums**

We can also choose to read only a subset of columns. This is done by setting the index_col option to a numeric value, which will lead to columns being read from 0 to whichever index we set the parse columns value to. We will set index_col=2. In this instance which reads the first 3 columns in the Excelfile.

```python
df = pd.read_excel('IMDB.xlsx',  index_col = 2)
df.head()
```

**Setting an index while reading data**

By default read_excel lables zeros with the numeric index, starting with 0. We can set our index or row lables to a higher value or to our choosing. To do this, we will pass a column name from our dataset to the index_col option.

```python
df = pd.read_excel('IMDB.xlsx',  index_col = 'Title')
df.head()
```
